```python
#Importar bibliotecas necesarias
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

#Cargar datos 
datos =  pd.read_csv('ecommerce_customer.csv')


#Seleccionar las variables (X) y el objetivo (y)
X = datos[['Customer Age', 'Total Purchase Amount', 'Product Price']]
y = datos['Churn']
```

```python
X_entrenamiento, X_prueba, y_entrenamiento, y_prueba = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
#Crear el modelo de regresin logstica

modelo = LogisticRegression()

#Entrenar el modelo
modelo.fit(X_entrenamiento, y_entrenamiento)
```

```python
#Predecir con los datos de prueba

y_prediccion = modelo.predict(X_prueba)

#Evaluar la precisin del modelo

precision = accuracy_score(y_prueba, y_prediccion)

print(f"La precisin del modelo es: {precision:.2f}")
```
